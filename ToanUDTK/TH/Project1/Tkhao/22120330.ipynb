### 1. Phép khử Gauss:
- Duyệt qua từng cột của ma trận A, từ cột đầu tiên đến cột cuối cùng:
  - Tìm hàng không chứa 0 tại vị trí cột hiện tại.
  - Nếu không tìm thấy, bỏ qua cột hiện tại.
  - Nếu tìm thấy, hoán đổi hàng chứa 0 với hàng đang xét và thực hiện phép toán để biến tất cả các phần tử bên dưới thành 0.
  - Lặp lại quá trình cho đến khi biến đổi hết các cột hoặc hàng của ma trận.
### 2. Phép thế ngược:
- Tính các nghiệm của hệ phương trình từ ma trận tam giác trên bằng phép thế ngược:
  - Tính hạng của ma trận A và ma trận mở rộng (A|B) để xác định trường hợp nghiệm:
    - Nếu rank(A) != rank(A|B) Hệ vô nghiệm => Kết thúc
    - Nếu rank(A) = rank(A|B) < n Hệ có vô số nghiệm
    - Còn lại Hệ có nghiệm duy nhất
  - Tìm nghiệm dựa trên trường hợp đã xác định:
    - Từ hàng cuối cùng của ma trận, tính giá trị của biến cuối cùng.
    - Dùng giá trị của biến đã tính được để tính các biến khác từ dòng trên.
  - In nghiệm và kết thúc chương trìnhn nghiệm và kết thúc chương trìnhIn nghiệm và kết thúc chương trìnhính các biến khác từ dòng trên.
  - In nghiệm và kết thúc chương trình

```python
import numpy as np
from sympy import symbols
import string

#Kiểm tra dòng row có tất cả các phần tử là 0 hay không
def isZeroRow(matrix, row):
    return all(element == 0 for element in matrix[row])

#Thêm dòng có các phần tử = 0 để ma trận có kích thước hợp lệ
def makeMatrix(matrix):
    rows, cols = matrix.shape
    max_dim = cols - 1;

    square_matrix = np.zeros((max_dim, cols))
    square_matrix[:rows, :cols] = matrix
    return square_matrix

#In ma trận
def display_matrix(matrix):
    for row in matrix:
        for element in row:
            if np.isclose(element, 0):
                element = 0
            print(f"{element:.2f}".rjust(6," "), end=" ")
        print()

#Phép khử Gauss
def Gauss_elimination(matrix):
    # Xác định số hàng và số cột của ma trận
    rows, cols = matrix.shape
    result_matrix = np.copy(matrix)
    
    pivot_row = 0
    for col in range(cols):
        nonzero_row = pivot_row

         # Tìm hàng không bằng 0 trong cột hiện tại
        while nonzero_row < rows and result_matrix[nonzero_row, col] == 0:
            nonzero_row += 1

        # Nếu không tìm thấy hàng không bằng 0, tiếp tục sang cột tiếp theo
        if nonzero_row == rows:
            continue
            
        # Hoán đổi dòng pivot với dòng không bằng 0
        result_matrix[[pivot_row, nonzero_row]] = result_matrix[[nonzero_row, pivot_row]]

        # Lấy giá trị của dòng pivot ở cột hiện tại
        pivot_value = result_matrix[pivot_row, col]
        # Chuẩn hóa dòng pivot bằng cách chia cho giá trị của pivot
        result_matrix[pivot_row] /= pivot_value

        for i in range(pivot_row + 1, rows):
            multiplier = result_matrix[i, col]
             # Loại bỏ các phần tử dưới pivot bằng cách lấy hàng hiện tại trừ đi một bội số của hàng pivot
            result_matrix[i] -= multiplier * result_matrix[pivot_row]
        
        pivot_row += 1

    return result_matrix

#Phép thế ngược
def back_substitution(matrix):
    # Xác định số hàng và số cột của ma trận
    rows, cols = matrix.shape
    # Khởi tạo vector nghiệm x với số phần tử bằng số cột trừ đi 1 trừ cột ma trận B
    x = np.zeros(cols-1)
    # Lấy ma trận A bằng cách loại bỏ cột cuối cùng của ma trận đầu vào
    Amatrix = matrix[:, :-1]
    # Tính rank của ma trận (A|B) và ma trận A
    ABrank = sum(1 for row in matrix if any(row))
    Arank = sum(1 for row in Amatrix if any(row))
    
    if Arank != ABrank:
        print("Hệ phương trình vô nghiệm.")
        return
    elif Arank == ABrank and Arank < cols - 1:
        freeVar = []
        index = 0
        alphabet = symbols(list(string.ascii_lowercase)) # Mảng các kí tự dùng làm ẩn tự do
        # Xếp lại ma trận bậc thang đầu vào sao cho các số dẫn đầu nằm đúng hàng của nó
        for i in range(rows):
            for j in range(cols):
                while (isZeroRow(matrix, i)==False):  
                    if matrix[i][j] != 0:
                        if i != j:
                            curCol = j
                            temp = np.array(matrix[curCol])
                            matrix[curCol]=matrix[i]
                            matrix[i] = temp
                    break
        # Duyệt theo đường chéo để đánh dấu ẩn tự do
        for i in range(rows):
            if all(matrix[j][i] == 0 for j in range(i,rows)):
                freeVar.append(alphabet[index])
                print(f"x_{i+1} là biến tự do")
                index += 1
            else:
                freeVar.append(0)
            
        x = np.array(freeVar)

        # Nhân tích vô hướng giữa vector nghiệm và hàng tường ứng để cho ra biểu thức tổng quát
        for i in range(rows-1,-1,-1):
            if matrix[i][i] == 0:
                continue
            s = x.dot(Amatrix[i])
            x[i] = (matrix[i][cols-1] - s) / matrix[i][i]
        # In nghiệm
        print("Nghiệm tổng quát của hệ phương trình:")
        for i in range(rows):
            if(str(x[i]).isalpha() and len(str(x[i]))==1):
                print(f"x_{i+1} = {x[i]} ∈ R")
            else:
                print(f"x_{i+1} = {x[i]}")
        return

    # Nếu không phải 2 trường hợp trên hệ có nghiệm duy nhất
    print("Hệ phương trình có nghiệm duy nhất.")
    for i in range(rows-1,-1,-1):
        x[i] = matrix[i][-1] / matrix[i][i]
        for j in range(i - 1, -1, -1):
            matrix[j][-1] -= matrix[j][i] * x[i]
    
    for i in range(rows):
        print(f"x_{i+1} = {x[i]:.2f}")
        
matrices = [
    #Có nghiệm duy nhất
    np.array([
        [1, 3, 1, -1, 2, 12],
        [-2, -9, -5, 11, -13, -48],
        [4, 13, 7, -7, 19, 56],
        [-2, -7, -2, 8, -12, -47],
        [0, -1, -2, -2, 11, 20]
    ]),
    #Vô số nghiệm 1 ẩn tự do
    np.array([
        [1, 3, -1, 2, 2, 0],
        [-1, 0, 4, -2, 1, -9],
        [4, 13, -2, 3, 12, -6],
        [3, 11, 0, 1, 11, -9],
        [1, 2, -1, -4, 3, 8]
    ]),
    #Vô số nghiệm nhiều ẩn tự do
    np.array([
        [1, -3, -5, -2, -2, 0],
        [-2, 6, 11, 5, 3, 0],
        [-3, 9, 15, 6, 3, 0],
        [1, -3, -10, -7, 6, 0],
        [-1, 3, 5, 2, 5, 0]
    ]),
    #Vô nghiệm
    np.array([
        [1, 0, 0, -2, -4, -1],
        [5, 3, -3, -22, -20, 7],
        [2, 1, -2, -8, -8, 5],
        [0, -3, -1, 17, 15, -15],
        [-2, -3, -1, 12, -4, 5]
    ])
]

for idx, matrix in enumerate(matrices, start=1):
    print(f"Ma trận {idx}:")
    matrix = makeMatrix(matrix)
    result = Gauss_elimination(matrix)
    print("Dạng bậc thang của ma trận:")
    display_matrix(result)
    back_substitution(result)
    print()
```
