## Giải thuật:

### 1. Phép khử gauss
#### Bước 1: Xác định cột trái nhất không chứa toàn số 0: Duyệt qua từng cột để tìm cột đầu tiên không chứa toàn số 0.
#### Bước 2: Đổi chỗ hai dòng để đưa số hạng khác 0 ở dưới về đầu cột: Sử dụng phép đổi chỗ hai dòng để đưa số hạng khác 0 về trên cùng của cột.
#### Bước 3: Chuẩn hóa dòng đầu để có số dẫn đầu là 1: Nếu số dẫn đầu của hàng đầu không phải là 1, thực hiện phép chuẩn hóa dòng để có số dẫn đầu là 1.
#### Bước 4: Biến các số hạng bên dưới số dẫn đầu thành 0: Sử dụng phép cộng và nhân để biến các số hạng dưới số dẫn đầu thành 0.
#### Bước 5: Lặp lại các bước cho đến khi được ma trận bậc thang: Lặp lại quá trình trên cho tất cả các cột cho đến khi ma trận trở thành dạng bậc thang.

### 2.Phép thế ngược:
#### Bước 1: Bắt đầu từ hàng dưới cùng (hàng cuối cùng): Với mỗi biến, bắt đầu từ dòng cuối cùng của ma trận đã được biến đổi theo phép Gauss.
#### Bước 2: Thay thế giá trị của biến vào các dòng phía trên: Sử dụng giá trị của biến đã được xác định để thay vào các dòng phía trên và tính toán giá trị của các biến còn lại.
#### Bước 3: Lặp lại cho tất cả các biến: Tiếp tục quá trình trên cho tất cả các biến, bắt đầu từ dòng cuối cùng và đi lên cho đến khi tất cả các biến đã được xác định.

```python
def gauss_jordan(A, B):
    n = len(A)
    M = [[0] * (n + 1) for _ in range(n)]

    for i in range(n):
        for j in range(n):
            M[i][j] = A[i][j]
        M[i][n] = B[i]

    for i in range(n):
        non_zero_row = -1
        for k in range(i, n):
            if M[k][i] != 0:
                non_zero_row = k
                break

        if non_zero_row != -1:
            M[i], M[non_zero_row] = M[non_zero_row], M[i]

        pivot = M[i][i]

        if pivot == 0:
            continue

        for j in range(i, n + 1):
            M[i][j] /= pivot

        for j in range(n):
            if i != j:
                factor = M[j][i]
                for k in range(i, n + 1):
                    M[j][k] -= factor * M[i][k]

    rank_A = n
    rank_M = n

    for i in range(n):
        all_zero = all(M[i][j] == 0 for j in range(n))
        if all_zero:
            rank_A -= 1

    for i in range(n):
        if M[i][n] == 0:
            all_zero = all(M[i][j] == 0 for j in range(n))
            if all_zero:
                rank_M -= 1

    if rank_A == rank_M == n:
        print("Hệ phương trình có một nghiệm duy nhất:")
        solutions = [M[i][n] for i in range(n)]
        for i, sol in enumerate(solutions):
            print(f"x[{i + 1}]: {sol}")
    elif rank_A < rank_M:
        print("Hệ phương trình vô nghiệm.")
    else:
        print("Hệ phương trình có vô số nghiệm.")

# Dữ liệu đầu vào
A = [[1, 3, 1, -1, 2],
     [-2, -9, -5, 11, -13],
     [4, 13, 7, -7, 19],
     [-2, -7, -2, 8, -12],
     [0, -1, -2, -2, 11]]

B = [12, -48, 56, -47, 20]

gauss_jordan(A, B)
```
