<h5><b>PHẦN I. Phép khử Gauss (Gauss Elimination)</b></h5>
<ol>
    <li>
        <div><b>Yêu cầu.</b> Viết hàm Gauss_elimination(A), trong đó: </div>
        <div>   
        <ul style = "margin-left: 25px">
            <li>Input: A là ma trận mở rộng của hệ phương trình</li>
            <li>Output: Ma trận bậc thang có được từ ma trận A</li>
        </ul>
        </div>
    </li>
    <li>
        <div><b>Giải thuật.</b></div>
        <div style = "margin-left: 25px">
            <div><b>B1. </b>Xác định cột trái nhất không chứa toàn số 0.</div>
            <div><b>B2. </b>Chọn dòng có số hạng dẫn đầu có giá trị tuyệt đối lớn nhất và hoán đổi vị trí với dòng đang xét pivot (nếu cần)</div>
            <div><b>B3. </b>Với lần lượt những hàng bên dưới: tìm hệ số k = số hạng dẫn đầu / pivot, sau đó trừ dòng đi k lần dòng pivot để số hạng bên dưới số dẫn đầu (pivot) bằng 0</div>
            <div><b>B4. </b>Lặp lại các bước cho những dòng dưới đến khi được ma trận bậc thang</div>
        </div>
    </li>
    <li>
        <div><b>Cài đặt thuật toán.</b></div>
    </li>
</ol>



```python
import numpy as np
from fractions import Fraction

def Gauss_elimination(A):
    # Xác định kích thước của ma trận 
    m, n = A.shape

    # Khởi tạo pRow và pCol dùng để lưu vị trí của pivot
    pRow = 0
    pCol = 0

    # Khởi tạo ma trận X có kích thước nxn dùng để đánh dấu ẩn tự do
    # Mặc định tất cả các ẩn đều là ẩn tự do
    X = np.full((n, n), Fraction(0))

    for i in range(0, n - 1):
        X[i, i] = 1 

    while pRow < m and pCol < n:
        # Tìm phần tử có trị tuyệt đối lớn nhất trong cột pCol
        imax = pRow

        for i in range(pRow + 1, m):
            if abs(A[imax, pCol]) < abs(A[i, pCol]):
                imax = i

        # Nếu phần tử có trị tuyệt đối lớn nhất bằng 0 thì chứng tỏ cột đang xét chứa 
        # toàn số 0 nên ta sẽ xét cột tiếp theo. Đồng thời khi đó nếu hệ  phương trình
        # có vô số nghiệm thì x_i sẽ là ẩn tự do
        if A[imax, pCol] == 0:
            pCol += 1
        else:
            X[pCol, pCol] = 0
            
            # Hoán đổi vị trí 2 dòng (nếu 2 dòng khác nhau)
            if imax != pRow:
                A[[imax, pRow],] = A[[pRow, imax],]

            # Trừ dòng bên dưới một số lần dòng pRow để được số dẫn đầu là 0
            for i in range(pRow + 1, m):
                k = A[i, pCol] / A[pRow, pCol]
                A[i,] = A[i,] - k * A[pRow,]

            pRow += 1
            pCol += 1

    return A, X
```

```python
# Nhập vào ma trận mở rộng
A = np.matrix([[1, 5, 1, -3, 4, -7],
[-2, -9, 1, 2, -6, 1],
[2, 5, -16, 14, 4, 57],
[1, 2, -9, 9, 2, 34],
[0, -2, -4, 8, -5, 22]], dtype = object)

# Xác định kích thước ma trận
m, n = A.shape

# Chuyển các phần tử của ma trận sang cấu trúc phân số để hạn chế sai sót do sai số
for i in range(m):
    for j in range(n):
        A[i,j] = Fraction(A[i,j])

# Thực hiện Gauss-elimination và in ra ma trận bậc thang
A, X = Gauss_elimination(A)

# In ma trận bậc thang
print("Ma trận bậc thang:")

for i in range(m):
    for j in range(n):
        x = str(A[i, j])
        x = ' ' * (10 - len(x)) + x
        print(x, end = '')
    print()
```

Ví dụ, với hệ phương trình có 4 ẩn x_0, x_1, x_2, x_3 thì sẽ được lưu vào ma trận x có kích thước 5x5: x_2 là ẩn tự do, khi đó dòng 2 sẽ là [0, 0, 1, 0, 0]; với x_0 = 2 * x_2 + 10 thì dòng 0 là [0, 0, 2, 0, 10]</div>

```python
# Hàm này được sử dụng để biểu diễn x và in ra
def RepresentX(X):
    m, n = X.shape 
    
    for i in range(0, n - 1):
        if X[i, i] == 1:
            print(f'\tx{i + 1} ∈ R')
        else:
            x = ''
                
            for j in range(i + 1, n - 1):
                if X[i,j] != 0:
                    if x != '' and X[i, j] > 0:
                        x += '+'
                        
                    x += f'{X[i, j]}*x{j + 1}'
                
            if X[i, n - 1] != 0: 
                if X[i, n - 1] > 0 and x != '':
                    x += '+'

                x += str(X[i, n - 1])
                
            if x == '':
                print(f'\tx{i + 1} = 0')
            else:
                print(f'\tx{i + 1} = {x}')
```

```python
def Back_substitution(A, X):
    m, n = A.shape;

    # Kiểm tra hệ phương trình có vô nghiệm hay không
    for i in range(m):
        zeroRow = True

        for j in range(n - 1):
            if A[i, j] != 0:
                zeroRow = False

        if zeroRow and A[i, n-1] != 0:
            print("Hệ phương trình vô nghiệm!")
            return

    # Trong trường hợp hệ phương trình có nghiệm
    infiniteSolution = False 
    X[n-1, n-1] = -1

    #Tìm hạng của ma trận
    rank = m

    for i in range(m - 1, -1, -1):
        zeroRow = True

        for j in range(0, n):
            if A[i, j] != 0:
                zeroRow = False

        if zeroRow:
            rank -= 1
    
    currRow = rank

    for i in range(n - 2, -1, -1):
        # Nếu có ẩn tự do, đánh dấu hệ phương trình có vô số nghiệm
        if X[i, i] == 1:
            infiniteSolution = True
        else:
            currRow -= 1

            for j in range(i + 1, n):
                X[i,] = X[i,] - A[currRow, j] / A[currRow, i] * X[j,]

    if infiniteSolution:
        print("Hệ phương trình có vô số nghiệm:")
    else: 
        print("Hệ phương trình có nghiệm duy nhất:")

    RepresentX(X)
            
```
